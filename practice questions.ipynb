```python
###Write a program to check if a list is sorted or not.
here sorted means the given object(list or tuple or set) is in descending or in ascending order(either one)
```

```python
###Create a dictionary of 5 students with their marks and find the student with the highest marks
```

```python
a={'bhuvi':70,
  'kavi':80,
  'karthiga':100,
  'priya':90}
topper=''
highest_mark=0
for name,mark in a.items():
    if mark>highest_mark:
        highest_mark=mark
        topper=name
print(topper,':',highest_mark)
```

```python
a={'bhuvi':700,
  'kavi':804,
  'karthiga':1000,
  'priya':903}
topper=''
top_score=0
for name,mark in a.items():
    if mark>top_score:
        top_score=mark
        topper=name
print(topper)
print(top_score)
```
