```python
### Recursive function
-function that calls itself
it's uses:
-we can reduce the length of the code and improves reusability
-complete the problems in easier way
```

```python
def factorial(n):
    b=1
    if n==0:
        print(1)
    for i in range(1,n+1):
        b=b*i
    print(b)
```

```python
factorial(5)
```

```python
def factorial(n):
    if n==0:
        b=1
    else:
        b=n*factorial(n-1)
    return b #most important=we should give return here then only recursive will take place
```

```python
factorial(5)
```

```python
#Keep only even numbers
num=[1,2,3,4,5,6,7,8]
result=list(filter(lambda x:x%2==0,num))
print(result)
```

```python
#Keep words longer than 3 letters
word=['bhuvi','python','hi','bye']
result=list(filter(lambda w:len(w)>3,word))
print(result)
```

```python
#Keep only positive numbers
a=[-1,-4,-5,3,2,2,1]
b=list(filter(lambda x:x>=0,a))
print(b)
```

```python
#filter() with a normal function (not lambda)
def even(x):
    return x%2==0
num=[2,3,4,4,5,6,7]
a=list(filter(even,num))
print(a)
```

MAP|

```python
-map() transforms data
-IT KEEPS THE LIST SIZE SAME
-It works element by element
-Needs list() to see result
-Mostly used with lambda
When SHOULD you use map()?
✔ When same operation is applied to all elements
✔ When function is small
✔ When code readability matters
```

```python
#Square each number
a=[1,2,3,4,5]
b=list(map(lambda x:x**2,a))
print(b)
```

```python
#Add 1 to every number
a=[1,2,3,4,5,6,7]
b=list(map(lambda x:x+1,a))
print(b)
```

```python
#Convert strings to uppercase
a=['bhuvi','python','guido']
b=list(map(lambda x:x.upper(),a))
print(b)
```

```python
#map() with a normal function (not lambda)
def square(x):
    return x*x
num=[2,3,4,5]
a=list(map(square,num))
print(a)
```

REDUCE

```python
->reduce() takes many values and reduces them to ONE single value.
->reduce() is not built-in in Python
->You must import it before using it
     VERY IMPORTANT
->The function inside reduce() must take TWO arguments

map → change
filter → choose
reduce → combine


nums
 ↓
filter()   → gives filtered values
 ↓
map()      → gives transformed values
 ↓
reduce()   → combines values into ONE result
```

```python
num=[1,2,3,4,5]
a= reduce(lambda a,b:a+b,num)
print(a)
```

```python
#Sum of squares of even numbers
num=[1,2,3,4,5,6] #Use list() to SEE data 
                  #Skip list() when PASSING data
even=filter(lambda x:x%2==0,num)
square=map(lambda x:x**2,even)
result=reduce(lambda x,y:x+y,square)
print(result)
```

```python
#Sum of squares of even numbers
num=[1,2,3,4,5,6] #Use list() to SEE data 
                  #Skip list() when PASSING data
even=filter(lambda x:x%2==0,num)
square=map(lambda x:x**2,even)
result=reduce(lambda x,y:x+y,square)
print(result)
```

```python
#Product of positive numbers after doubling them
nums = [-2, 3, -1, 4, 5]
positive=filter(lambda x:x>0,nums)
doubling=map(lambda x:x*2,positive)
result=reduce(lambda x,y:x*y,doubling)
print(result)
```

```python
#Total length of words longer than 3 characters
letter=['bhuvi','hi','python','hey','guido']
leng=filter(lambda x:len(x)>3,letter)
lengt=map(lambda x:len(x),leng)
charac=reduce(lambda x,y:x+y,lengt)
print(charac)
```

```python
#Sum of cubes of odd numbers
num=[1, 2, 3, 4, 5]
odd=filter(lambda x:x%2!=0,num)
cube=map(lambda x:x**3,odd)
result=reduce(lambda x,y:x+y,cube)
print(result)
```
